
````java
//abans
int[] array = {4,2,7,9,1,2,6,-1,8};
println(Arrays.toString(array)); //imprimirà [4, 2, 7, 9, 1, 2, 6, -1, 8]

//TODO
int big = array[0];
int low = array[0];
int bigpos = 0;
int lowpos = 0;

println(Arrays.toString(array)); //imprimirà [4, 2, 7, 9, 1, 2, 6, -1, 8]
for(int index =0; index<array.length;index++){
       if(big > array[index]){//buscamos el num más grande
        big = array[index];
        bigpos =index;}//guardamos su posicion
    
        if(low < array[index]){//buscamos el num más pequeño
          low = array[index];
          lowpos = index;//guardamos su posicion
        }    
}
//cambiamos posiciones
array[lowpos] = big;
array[bigpos] = low;


//després
println(Arrays.toString(array));


//després
println(Arrays.toString(array)); //imprimirà [4, 2, 7, -1, 1, 2, 6, 9, 8]
````
